```python
#display the 18 years in order of most to least death counts per year of all causes
most_deaths = df.nlargest(18, 'Deaths')
print(most_deaths)

plt.figure(figsize = (10,6))
plt.bar(most_deaths['Year'], most_deaths['Deaths'], color = 'blue')
plt.title('Highest Number of Deaths Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.show()
```

*Here we can see that in the US over the last 18 years there has been a consistent upward trend in the number of deaths from around 2010 to 2017, in contrast to the relative stable death count in the preceding years*

*Now lets look at California*

```python
#creates condition for a certain state's row details
desired_state = 'California'
desired_cause = 'All causes'
filtered_rows_state = df[(df['State'] == desired_state) & (df['Cause Name'] == desired_cause)]
print(filtered_rows_state)

#makes rows in ascending order
filtered_rows_state = filtered_rows_state.sort_values('Year')

#create line graph
plt.figure(figsize = (10,6))
plt.plot(filtered_rows_state['Year'], filtered_rows_state['Deaths'], marker = 'o', linestyle = '-', color = 'blue')

plt.title('Trend of Deaths in California for All Causes')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.grid(True)
plt.show()
```

*Here we can see that there has been an exponential growth in the number of deaths per year in California. Growing from just under 230,000 deaths in 1999 to just under 270,000 deaths in 2017. This grwoth is prominent from around 2010 and on.*

*Now lets compare these graphs*

```python
#conditional states and cause of death
states = ['California', 'Texas', 'Florida']
desired_cause = 'All causes'

#create empty dataframe for rows from selected states
filtered_rows_all_states = pd.DataFrame()

#loop each state and move conditional rows to the dataframe
for state in states:
    filtered_rows_state = df[(df['State'] == state) & (df['Cause Name'] == desired_cause)]
    filtered_rows_all_states = pd.concat([filtered_rows_all_states, filtered_rows_state])
    
#sort by year
filtered_rows_all_states = filtered_rows_all_states.sort_values('Year')
    
plt.figure(figsize = (10,6))

#loop for plotting
for state in states:
    filtered_rows_state = filtered_rows_all_states[filtered_rows_all_states['State'] == state]
    plt.plot(filtered_rows_state['Year'], filtered_rows_state['Deaths'], marker = 'o', linestyle = '-', label = state)
        
plt.title('Trend of Deaths for All Causes in Different States')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.grid(True)
plt.legend()
plt.show()
```

*Here we can see all three states together on one graph, we can see some similarities overall with an upward trend in death count (which is natural due to an increase in population). We can however, see a very specific similarity between Texas and Florida from around 2012 to 2015 where the growth and numbers were nearly identical.*
